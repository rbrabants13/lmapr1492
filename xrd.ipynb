```python
from pymatgen.analysis.diffraction.xrd import *
from pymatgen.analysis.diffraction.core import *
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
with MPRester ("VTUv2zAwLlpXxhqy") as m:
    #On définit les variables nécessaires pour la suite
    structure = m.get_structure_by_material_id("mp-570485")
    ConventionalStructure = SpacegroupAnalyzer(structure).get_conventional_standard_structure()
    xrd = XRDCalculator(1.54060) #On crée un objet XRD Calculator avec une longueur d'onde lamda de 1.54060 Å
    pattern = xrd.get_pattern(ConventionalStructure) #renvoie un object DiffractionPattern
```

```python
#L'objet DiffractionPattern se trouve dans le module pymatgen.analysis.diffraction.core

theta = pattern.x #On crée un tableau contenant les angles 2*theta
intensite=pattern.y #On crée un tableau contenant les intensités en %
hkl = pattern.hkls #On crée un tabeau de tableaux contenant tous les indices hkl 
```

```python
#On reprend les 3 premiers pics et on les affiche

print("\n Le premier pic est observé avec un angle 2θ =",theta[0],"°, il correspond au plan",hkl[0][0]["hkl"],"et a comme intensité",intensite[0],"%")
print("\n Le second pic est observé avec un angle 2θ =",theta[1],"°, il correspond au plan",hkl[1][0]["hkl"],"et a comme intensité",intensite[1],"%")
print("\n Le troisième pic est observé avec un angle 2θ =",theta[2],"°, il correspond au plan",hkl[2][0]["hkl"],"et a comme intensité",intensite[2],"%")
```

```python
#Pour voir si ce qu'on a fait ressemble aux données trouvées sur MP, on plot le graphique
print("\n On affiche le graphique globale de la diffraction aux rayons X ")
xrd.show_plot(ConventionalStructure)
```
